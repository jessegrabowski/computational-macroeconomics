```python
import sympy as sp
import numpy as np
```

We consider the model:

$$ x_t - \theta x_{t-1} = \epsilon_t - \phi \epsilon_{t-1}$$

With $\epsilon_t \sim N(0, 1)$.

# Simulation via Numpy

## Steady State

We can use sympy to do the symbolic manipulations for us. Unsurprisingly, the answer is 0.

```python
x, e, theta, phi = sp.symbols('x e theta phi')

# in steady-state, x_t = x_{t-1} = x
eq = sp.Eq(x - theta * x, e - phi * e)
# in the non-stochastic steady-state, e = 0

eq = eq.subs({e:0})
sp.solve(eq, x)[0]
```

## Simulation

We can cast the problem into state-space form for easy simulation:

$$a_t = Ta_{t-1} + R\nu_t$$

Where: 
$$\begin{align}a_t &= \begin{bmatrix} x_t \\ \epsilon_t \end{bmatrix} \\
               T &= \begin{bmatrix} \theta & -\phi \\ 0 & 0 \end{bmatrix} \\
               R &= \begin{bmatrix} 1 \\ 1 \end{bmatrix} \\
               \nu_t &= \begin{bmatrix} \epsilon_t \end{bmatrix}
               \end{align}$$ 

```python
def simulate_ARMA11(ϕ=0.4, θ=0.4, x0=None, sim_length=200, seed=None):
    rng = np.random.default_rng(seed)
    if x0 is None:
        x0 = np.zeros((2, 1))
    
    T = np.array([[θ, -ϕ],
                  [0,  0]])
    
    R = np.array([1, 1]).reshape(-1, 1)
    
    innovations = rng.normal(size=sim_length)
    output = np.empty((sim_length, 2, 1))
    output[0] = x0
    for t in range(1, sim_length):
        output[t] = T @ output[t-1] + R * innovations[t]
    
    return output.squeeze()
```

### Default Parameters

With parameters $\theta = \phi = 0.4$, the system has very little persistence. As a result, the series $x_t$ very closely resembles the innovations.

```python
import matplotlib.pyplot as plt
seed = sum(map(ord, 'ARIMA(11) simulation'))

def simulate_and_plot(ϕ=0.4, θ=0.4, x0=None, sim_length=200, seed=None):
    fig, ax = plt.subplots(2, 1, figsize=(14, 4), dpi=144, sharex=True)
    data = simulate_ARMA11(ϕ=ϕ, θ=θ, x0=x0, sim_length=sim_length, seed=seed)
    names = ['$x_t$', r'$\epsilon_t$']
    for axis, x, name in zip(fig.axes, data.T, names):
        axis.plot(x)
        axis.set_title(name, fontsize=16)
        [spine.set_visible(False) for spine in axis.spines.values()]
        axis.grid(ls='--', lw=0.5)
    fig.tight_layout()
    plt.show()
    
simulate_and_plot()
```

## Unit Root

By allowing for more persistence, setting $\theta = 0.99$, we can see a much different picture

```python
simulate_and_plot(θ=0.99)
```

## Large MA parameter

Increasing $\phi = 2.5$ does not have the same effect. Dynamics still resemble the underlying innovations, albeit scaled (note the y-axis differences between the two plots)

```python
simulate_and_plot(ϕ=2.5)
```

# Using gEconpy

It's possible to do this exercise in gEconpy as well. Here is the GCN file

```python
with open('arima11.gcn', 'r') as file:
    print(file.read())
```

## Load and solve model

```python
import gEconpy as ge
mod = ge.gEconModel('arima11.gcn')
```

```python
mod.steady_state()
mod.solve_model(model_is_linear=True)
```

## Steady State

It's zero, as promised

```python
mod.print_steady_state()
```

## Simulation

```python
def simulate_and_plot_geconpy(mod, ϕ=0.4, θ=0.4, sim_length=200):
    original_params = mod.free_param_dict.copy()
    new_params = {'phi':ϕ, 'theta':θ}
    mod.free_param_dict.update(new_params)
    mod.steady_state(verbose=False)
    mod.solve_model(model_is_linear=True, verbose=False)
    
    data = mod.simulate(n_simulations=1, 
                        simulation_length=sim_length,
                        shock_cov_matrix=np.eye(1)).T.droplevel(1)
    
    fig, ax = plt.subplots(2, 1, figsize=(14, 4), dpi=144, sharex=True)
    data.x.plot(ax=ax[0], title='$x_t$', legend=False)
    data.eps.plot(ax=ax[1], title='$\epsilon_t$', legend=False)
    [spine.set_visible(False) for axis in fig.axes for spine in axis.spines.values()]
    [axis.grid(ls='--', lw='0.5') for axis in fig.axes]
    fig.tight_layout()
    plt.show()
    
    # Restore orignal parameters
    mod.free_param_dict.update(original_params)
```

### Default Parameters

Results of simulation are qualitatively identical to those obtained "by hand".

```python
simulate_and_plot_geconpy(mod)
```

 ### Unit Root
 We can see the same changes when we change parameters as well

```python
simulate_and_plot_geconpy(mod, θ=0.99)
```

### Large MA term

```python
simulate_and_plot_geconpy(mod, ϕ=2.5)
```
